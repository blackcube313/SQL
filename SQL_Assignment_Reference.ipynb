## Q1 --- List all the directors who directed a 'Comedy' movie in a leap year. (You need to check that the genre is 'Comedy’ and year is a leap year) Your query should return director name, the movie name, and the year.

```python
%%time
# Write your sql query below

query = """
        SELECT DISTINCT p.Name AS Director,m.title, SUBSTR(m.year,-4) AS YEAR, g.Name AS Genre
        FROM Person p, Movie m, M_Director md, Genre g, M_Genre mg
        WHERE m.MID=mg.MID AND m.MID = md.MID AND g.Name like '%Comedy%' AND TRIM(md.PID)=p.PID AND mg.GID = G.GID
        AND (SUBSTR(m.year,-4)%4=0 AND SUBSTR(m.year,-4)%100!=0 OR SUBSTR(m.year,-4)%400=0) 
   
        
        
        """

q1 = pd.read_sql_query(query, conn)    
print(q1.shape)
q1
```

## Q4 --- List all directors who directed 10 movies or more, in descending order of the number of movies they directed. Return the directors' names and the number of movies each of them directed.

```python
%%time
# Write your sql query below

query = """
        SELECT DISTINCT TRIM(p.Name) AS 'Director', COUNT(*) Movie_Count
        FROM Person p
        JOIN M_Director md ON TRIM(p.PID) = TRIM(md.PID)
        GROUP BY TRIM(md.PID) HAVING COUNT(*) >= 10 ORDER BY Movie_Count DESC
        
        """
      
        

q4 = pd.read_sql_query(query, conn)    
print(q4.shape)
q4.head()
```

## Q5.a --- For each year, count the number of movies in that year that had only female actors.

```python
%%time
# Write your sql query below

query = """
       SELECT COUNT (m.MID) AS Num_Of_Movies, SUBSTR(m.year,-4) AS YEAR
       FROM Movie m
       WHERE m.MID NOT IN (
            SELECT m.MID
            FROM Movie m
            JOIN M_Cast mc
            ON mc.MID = m.MID
            JOIN Person p
            ON TRIM(mc.PID) = p.PID
            WHERE TRIM(p.Gender) = 'Male'
            
            UNION
            
            SELECT mc.MID 
            FROM M_Cast mc
            WHERE TRIM(mc.PID) IS NULL
            )
       
       GROUP BY SUBSTR(m.year,-4)
       ORDER BY SUBSTR(m.year,-4) ASC
       
       
     
       
        """

q5a = pd.read_sql_query(query, conn)    
print(q5a.shape)
q5a
```

## Q5.b --- Now include a small change: report for each year the percentage of movies in that year with only female actors, and the total number of movies made that year. For example, one answer will be: 1990 31.81 13522 meaning that in 1990 there were 13,522 movies, and 31.81% had only female actors. You do not need to round your answer.

```python
%%time
# Write your sql query below

query = """
        
       SELECT x.F_YEAR, ((x.Num_of_Movies*100)/(y.All_Movies*1.000000)) AS Percent, y.All_Movies AS 'Total_Movies'
        FROM
                (SELECT COUNT(m.MID) AS Num_Of_Movies, SUBSTR(m.year,-4) AS F_YEAR
                   FROM Movie m
                   WHERE m.MID NOT IN (
                        SELECT m.MID
                        FROM Movie m
                        JOIN M_Cast mc
                        ON TRIM(mc.MID) = m.MID
                        JOIN Person p
                        ON TRIM(mc.PID) = p.PID
                        WHERE TRIM(p.Gender) = 'Male'

                        UNION

                        SELECT mc.MID 
                        FROM M_Cast mc
                        WHERE TRIM(mc.PID) IS NULL
                        )

                   GROUP BY SUBSTR(m.year,-4)
                   ORDER BY SUBSTR(m.year,-4) ASC)
                
                x INNER JOIN
        
                (SELECT count(m.title) AS All_Movies, SUBSTR(m.year,-4) AS ALL_YEAR
                FROM Movie m
                GROUP BY SUBSTR(m.year,-4)
                ORDER BY SUBSTR(m.year,-4) ASC) y
                
                ON x.F_YEAR = y.ALL_YEAR


        
        """

q5b = pd.read_sql_query(query, conn)    
print(q5b.shape)
q5b
```

## Q6 --- Find the film(s) with the largest cast. Return the movie title and the size of the cast. By "cast size" we mean the number of distinct actors that played in that movie: if an actor played multiple roles, or if it simply occurs multiple times in casts, we still count her/him only once.

```python
%%time
# Write your sql query below

query = """
       SELECT m.title, COUNT(DISTINCT p.Name) AS 'Cast_Size'
       FROM Movie m
       JOIN M_Cast mc
       ON TRIM(m.MID) = mc.MID
       JOIN Person p
       ON TRIM(mc.PID) = p.PID
       
       
       GROUP BY m.title
       ORDER BY COUNT(DISTINCT p.Name) DESC
       LIMIT 1
        """

q6 = pd.read_sql_query(query, conn)    
print(q6.shape)
q6
```

## Q7 --- A decade is a sequence of 10 consecutive years. For example, say in your database you have movie information starting from 1965. Then the first decade is 1965, 1966, ..., 1974; the second one is 1967, 1968, ..., 1976 and so on. Find the decade D with the largest number of films and the total number of films in D.

```python
%%time
# Write your sql query below

query = """
        SELECT D.year AS Start, D.year+9 AS End, count(m.title) AS Total_Movies 
        FROM 
            (SELECT DISTINCT m.year from Movie m) AS D 
            JOIN Movie m
            ON m.year >= Start and m.year<= End 
            GROUP BY End 
            ORDER BY Total_Movies DESC 
            LIMIT 1
        
        
        
        
        """

q7 = pd.read_sql_query(query, conn)    
print(q7.shape)
q7.head()
```

## Q8 --- Find all the actors that made more movies with Yash Chopra than any other director.

```python
%%time
# Write your sql query below

query = """
        SELECT pa.Name AS ACTORS, pd.Name AS DIRECTOR, count(mc.MID) AS MOVIES_TOGETHER   
        FROM M_Cast mc
        JOIN M_Director md ON md.MID=mc.MID
        JOIN Person pa ON TRIM(mc.PID)=pa.PID
        JOIN Person pd ON TRIM(md.PID)=pd.PID
        WHERE pd.name LIKE '%yash%'
        GROUP BY mc.pid, md.pid 
        ORDER BY MOVIES_TOGETHER DESC
        
        """

q8 = pd.read_sql_query(query, conn)    
print(q8.shape)
q8.head()

```
